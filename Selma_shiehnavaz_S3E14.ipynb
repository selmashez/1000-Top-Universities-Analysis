```python
# Which factors contribute the most to the overall university score?

# let's make all columns numeric:
df['World Rank'] = pd.to_numeric(df['World Rank'], errors='coerce')
df['National Rank'] = pd.to_numeric(df['National Rank'], errors='coerce')
df['Research Output'] = pd.to_numeric(df['Research Output'], errors='coerce')
df['Quality Publications'] = pd.to_numeric(df['Quality Publications'], errors='coerce')
df['Influence'] = pd.to_numeric(df['Influence'], errors='coerce')
df['Citations'] = pd.to_numeric(df['Citations'], errors='coerce')
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
```

```python
columns_for_x_axis =['World Rank','National Rank','Quality of Education','Alumni Employment',
                        'Quality of Faculty','Research Output','Quality Publications','Influence' ]


for column in columns_for_x_axis:
    sns.lmplot(x=column, y='Score', data=df, aspect=1.5)
    plt.title(f'Score vs {column}')
    plt.show()
```

As is evident, the slopes of the lines are very close to each other. To precisely determine which factor affects the score, I will calculate the slope of each line.

```python
for column in columns_for_x_axis:
    
    x = df[column]
    y = df['Score']
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    print(f'The slope of the line for {Fore.RED}{column}{Style.RESET_ALL} vs Score is {Fore.BLUE}{slope:.3f}{Style.RESET_ALL}')
```

Based on the obtained values, the quality of the faculty has a greater impact on the university score compared to other numerical features.

```python
#What is the Probability that a Randomly Selected University is in North America and Has a Score Above 90?

count_of_North_America_and_score_above_90 = df[(df['Continent'] == "North America") & (df['Score'] >= 90)].shape[0]

p_intersection = count_of_North_America_and_score_above_90 / len(df)

print(f'the Probability that a Randomly Selected University is in North America and Has a Score Above 90 is {Fore.RED}{p_intersection * 100:.2f}{Style.RESET_ALL} percent')
```

```python
# Which countries have the highest average university scores?

average_scores = df.groupby("Location")['Score'].mean().reset_index()

average_scores.columns = ['Country', 'Average Score']

# Create the choropleth map
fig = px.choropleth(
    average_scores,
    locations='Country',              
    locationmode='country names',      
    color='Average Score',             
    hover_name='Country',              
    color_continuous_scale='Viridis',  
    title='Average University Score by Country'
)

fig.show()
```

We can see the average of university score for each country  using this plot and Switzerland is one of the countries that has high average score

```python
# Which Universities Have the Most Significant Gap Between "Quality of Education" and "Quality of Faculty

df['Difference_Faculty_Education'] = abs(df['Quality of Faculty'] - df['Quality of Education'])

df2 = df.copy()
df2.sort_values(by="Difference_Faculty_Education", ascending=False, inplace=True)
df2 = df2.head(5)

plt.figure(figsize=(12, 8))
sns.barplot(
    x='Difference_Faculty_Education', 
    y='Institution', 
    data=df2, 
    palette='viridis'
)
plt.xlabel('Difference Between Quality of Education and Quality of Faculty')
plt.ylabel('Institution')
plt.title('Top 5 Universities with the Largest Gap Between Education and Faculty Quality')
plt.show()
```
